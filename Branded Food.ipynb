**Baby Food time**

We are looking at category of 'Baby/Infant'

```python
baby_food = branded_foods[branded_foods["branded_food_category"]== 'Baby/Infant  Foods/Beverages']
baby_food.head()
```

```python
baby_food.brand_owner.value_counts()
```

```python
baby_food['ingredients'] = baby_food.ingredients.str.lower()
```

```python
babyfood_ingred = baby_food.ingredients.str.split(pat=",", expand=True).stack().value_counts()
```

```python
babyfood_ingred.head(50)
```

```python
remove_ing = ['organic', 'puree,', 'water', 'less', 'than', 'high', 'juice', 'ingredients:', '(organic', 'm.', 'c.', 'a', 'of:,']
```

```python
babyfood_ingred.drop(remove_ing, inplace = True, errors='ignore') 
```

```python
babyfood_df = babyfood_ingred.to_frame().reset_index() 
babyfood_df.head(20)
```

```python
babyfood_df.rename(columns = {'index':'ingredients', 0:'freq'}, 
          inplace = True)
```

```python
babyfood_df['ingredients'] = babyfood_df['ingredients'].str.replace('ingredients:', "")
```

```python
babyfood_df.columns
```

```python
babyfood_df.to_csv('babyfood_branded.csv')
```
